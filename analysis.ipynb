Clean the Data 

```python
# Data Cleaning: Standardize categorical values and parse dates
df["delivery_status"] = df["delivery_status"].str.strip().str.lower()
df["payment_method"] = df["payment_method"].str.strip().str.title()
df["order_date"] = pd.to_datetime(df["order_date"], format="%d-%m-%y")
```

```python
# after cleaning
df.head()      
```

```python
# Calculate key performance indicators (KPIs)
rows = len(df)
print(rows)
```

Total revenue

```python
# quantity often contains strings like "3", " 3 ", or "3 units"
df["quantity"] = (
    df["quantity"]
    .astype(str)
    .str.strip()
    .str.replace(",", "", regex=False)
)

df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
df["unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")
df["discount_applied"] = pd.to_numeric(df["discount_applied"], errors="coerce").fillna(0)

# Optional: remove rows where quantity or price couldn't be converted
df = df.dropna(subset=["quantity", "unit_price"])

```

```python
#calculates revenue for each order and extracts the month from the order date so the data can be analyzed or plotted by month
df["revenue"] = df["quantity"] * df["unit_price"] * (1 - df["discount_applied"])
df["revenue"].sum()
print(df["revenue"].sum())
```

```python
kpis = {
    "total_orders": df["order_id"].nunique(),
    "total_customers": df["customer_id"].nunique(),
    "total_revenue": df["revenue"].sum(),
    "avg_order_value": df.groupby("order_id")["revenue"].sum().mean(),
    "avg_discount": df["discount_applied"].mean(),
}

print(kpis)
```

```python
# Monthly Revenue Line Chart
monthly_revenue = df.groupby("month")["revenue"].sum().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Revenue by Region Bar Chart
revenue_by_region = df.groupby("region")["revenue"].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(revenue_by_region.index, revenue_by_region.values)
plt.title("Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
# Delivery Status Distribution Bar Chart
delivery_counts = df["delivery_status"].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(delivery_counts.index, delivery_counts.values)
plt.title("Delivery Status Distribution")
plt.xlabel("Delivery Status")
plt.ylabel("Order Count")
plt.tight_layout()
plt.show()
```
